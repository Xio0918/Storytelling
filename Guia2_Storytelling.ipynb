## **Justificación**

La justificación principal de este proyecto radica en la necesidad de entender con precisión la relación entre aspectos específicos de la personalidad y los patrones de consumo de sustancias legales e ilegales. El consumo de estas sustancias es una realidad compleja presente en diversas sociedades, y abordarla desde un enfoque basado en datos permite obtener perspectivas más profundas y objetivas.

Este estudio busca identificar y analizar patrones relevantes, explorando cómo factores como la impulsividad, la búsqueda de sensaciones, la amabilidad o la apertura a nuevas experiencias influyen en las decisiones relacionadas con el consumo de drogas. Al examinar esta conexión, podemos obtener información valiosa para la creación de intervenciones efectivas, programas de prevención y campañas de concientización fundamentadas en evidencias científicas.

Además, considerar múltiples drogas y diversas variables culturales, educativas y sociales proporciona una visión integral que facilita la comprensión de la complejidad detrás de las decisiones de consumo. El objetivo no es únicamente categorizar consumidores y no consumidores, sino revelar las causas subyacentes que guían estas decisiones.

## **Objetivo general:**

Identificar patrones en la relación entre atributos de personalidad y consumo de sustancias legales e ilegales mediante técnicas de clasificación, con el fin de predecir y evaluar perfiles de consumidores.

## **Objetivos específicos:**

1. Analizar y cuantificar la influencia de factores como la impulsividad, búsqueda de sensaciones y dimensiones del modelo NEO-FFI-R en el consumo de diferentes sustancias mediante modelos de clasificación.

2. Determinar la mejor estrategia de binarización para las categorías de consumo, optimizando la precisión en la identificación del perfil de consumidor o no consumidor.

3. Evaluar el riesgo potencial asociado al consumo de cada sustancia mediante el análisis predictivo basado en atributos de personalidad y datos demográficos.

## **1. Contextualización:**

La clasificación del tipo de consumidor de drogas basada en datos de personalidad es crucial debido al impacto significativo que el consumo de sustancias tiene sobre la salud pública, el bienestar social y la economía. Entender cómo características personales como la impulsividad, la búsqueda de sensaciones y otros factores demográficos y psicológicos influyen en el comportamiento de consumo, puede ayudar a desarrollar intervenciones preventivas y terapéuticas más efectivas. Los resultados servirán para diseñar programas de prevención y tratamiento adaptados específicamente a distintos perfiles psicológicos, educativos y culturales, contribuyendo así a reducir riesgos asociados al consumo de sustancias. Este análisis toma lugar en un contexto contemporáneo utilizando datos cuantificados provenientes de encuestas recientes realizadas a una población diversa y multicultural.

## **Descripción de la Base de Datos**

La base de datos proviene del estudio presentado en el artículo de E. Fehrman et al. (2015), titulado "The Five Factor Model of Personality and Evaluation of Drug Consumption Risk".

Esta base de datos contiene información sobre rasgos de personalidad y consumo de sustancias. Los datos fueron recopilados mediante una encuesta en línea, donde los participantes respondieron preguntas sobre su personalidad, impulsividad y uso de diversas sustancias.

## **Contenido de la Base de Datos**

**Rasgos de Personalidad (Big Five)**

Neuroticismo (nscore), Extraversión (escore), Apertura a la experiencia (oscore), Amabilidad (ascore) y Responsabilidad (cscore).


**Factores Psicológicos Adicionales**

Impulsividad (impulsive) y búsqueda de sensaciones (ss).


**Consumo de Sustancias**

Incluye el uso de alcohol, cafeína, nicotina, cannabis, cocaína, éxtasis, heroína, LSD, metanfetaminas, entre otras drogas legales e ilegales. Cada sustancia está cuantificada según la frecuencia de consumo, desde "Nunca usado" hasta "Usado en el último día".


**Origen de los Datos**

Los datos fueron recopilados a través de un cuestionario en línea basado en el modelo de los cinco grandes rasgos de personalidad (Big Five) y escalas de evaluación de consumo de drogas. Se utilizó un modelo de auto-reporte, por lo que los datos reflejan la percepción de los propios participantes.

### **Clúster de consumidores con K-Means**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Normalizar datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_encoded[["nscore", "escore", "oscore", "ascore", "cscore"] + consumo_cols])

# Aplicar K-Means con 3 grupos
kmeans = KMeans(n_clusters=3, random_state=42)
df_encoded["Cluster"] = kmeans.fit_predict(df_scaled)

# Visualizar los clusters en función de dos variables
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_encoded, x="nscore", y="cannabis", hue="Cluster", palette="viridis")
plt.title("Segmentación de Consumidores en Base a Personalidad y Consumo")
plt.show()
```

**Análisis de la Segmentación de Consumidores en Base a Personalidad y Consumo de Cannabis**

Tres grupos diferenciados (clusters 0, 1 y 2): Los consumidores han sido agrupados según su nivel de neuroticismo (nscore) y consumo de cannabis.

Cluster 0 (morado): Bajo consumo: Mayor concentración en los niveles más bajos de consumo de cannabis.

Cluster 1 (azul): Consumo moderado: Distribuido en todos los niveles de consumo con tendencia a valores intermedios de neuroticismo.

Cluster 2 (amarillo): Alto consumo: Se encuentra en los valores más altos de consumo y muestra dispersión en el neuroticismo.

### **Comparación del consumo de drogas por país**

```python
# Lista de sustancias a analizar
sustancias = ["alcohol", "cannabis", "coke", "nicotine", "meth", "heroin"]

# Lista de países únicos en la base de datos
paises = df["country"].unique()

# Recorrer cada país
for pais in paises:
    df_pais = df[df["country"] == pais]  # Filtrar datos por país

    # Crear figura
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
    fig.suptitle(f"Distribución del Consumo de Sustancias en {pais}", fontsize=16)

    # Recorrer sustancias y asignarlas a los subplots
    for i, sustancia in enumerate(sustancias):
        ax = axes[i // 3, i % 3]  # Posicionar en la matriz de subplots
        
        # Obtener conteo de categorías de consumo
        conteos = df_pais[sustancia].value_counts()
        
        # Crear gráfico de torta
        wedges, texts, autotexts = ax.pie(
            conteos.values, labels=None, autopct='%1.1f%%', 
            colors=plt.cm.Paired.colors, startangle=90, wedgeprops={'edgecolor': 'black'}
        )
        
        # Agregar leyenda fuera del gráfico
        ax.legend(wedges, conteos.index, title=sustancia.capitalize(), loc="center left", bbox_to_anchor=(1, 0.5), fontsize=9)
        ax.set_title(f"Consumo de {sustancia.capitalize()} en {pais}")

    # Ajustar el diseño
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```
